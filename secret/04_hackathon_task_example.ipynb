## Development stage

Start the tracker

```python
from codecarbon import EmissionsTracker
tracker = EmissionsTracker()
tracker.start()
```

Load the networks and construct onion layer feature vectors.

```python
import networkx as nx
import os, glob
import pandas as pd
```

```python
rows = []
targets = []

labels = os.listdir("data/train")
for label in labels:
    print(label)
    for path in glob.glob("data/train/" + label + "/*"):
        print(path)
        G = nx.read_adjlist(path, nodetype=int)
        od = nx.onion_layers(G)
        layer_size = {}
        for i in range(1, 101):
            layer_i = [v for v in od.keys() if od[v] == i]
            layer_size[i] = len(layer_i)
        print(layer_size)

        rows.append(pd.Series(layer_size, name=path))
        targets.append(label)

X_train = pd.DataFrame(rows)
y_train = pd.Series(targets)
```

Make a train/validation split in the data we have available currently.

```python
from sklearn.model_selection import train_test_split
(X_train_0, X_val_0, y_train_0, y_val_0) = train_test_split(X_train, y_train, train_size=0.5, random_state=33)
```

Let's try a random forest classifier. 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_0, y_train_0)
```

Predictions on the validation dataset:

```python
pred = rfc.predict(X_val_0)
print(pred)
```

The correct classes for the validation dataset:

```python
print(list(y_val_0))
```

Calculate accuracy.

```python
accuracy_0 = (pred == y_val_0).mean()
print(accuracy_0)
```

Looks good - let's finish development here.

```python
emissions_dev = tracker.stop()
print(emissions_dev)
```
